```python
import nltk
import pandas as pd

nltk.download('reuters')
from nltk.corpus import reuters
from nltk.stem.porter import *
from nltk.stem import WordNetLemmatizer

cats = reuters.categories()
print("Reuters has %d categories:\n%s" % (len(cats), cats))

fileids = reuters.fileids()
```

```python
nltk.download('stopwords')
nltk.download('omw-1.4')
nltk.download('wordnet')
```

```python
categories = []
text = []

for file in fileids:
    categories.append(reuters.categories(file))
    text.append(reuters.raw(file))

raw_df = pd.DataFrame({'ids':fileids, 'categories':categories, 'text':text})
raw_df
```

```python

df = pd.DataFrame()
df['ids'] = raw_df['ids'][:50]
df['categories'] = raw_df['categories'][:50]
df['text'] = raw_df['text'][:50]
```

```python
stop_list_nltk = nltk.corpus.stopwords.words('english')
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
def preprocessing(text):
    tk = text.lower()
    tk = re.split(r'[\s!$@#,.;]+',tk)
    
    n_tk = []
    for i in tk:
      if i not in stop_list_nltk:
        n_tk.append(i)
    tk_final = [lemmatizer.lemmatize(w) for w in n_tk]
    return tk_final 

```

```python
for i,t in enumerate(df['text']):
  df['text'][i] = preprocessing(t)
  
```

```python
for i,t in enumerate(df['text']):
  df['text'][i] = " ".join(t)
df
```

```python

frase = " ".join(df['text'])
frase
```

```python
tokens = frase.split()
vocab = sorted(set(tokens))
```

```python
vocab
```

bow de contagem



```python
bow_contagem = {}

for i,doc in enumerate(df['text']):
    bow_contagem[i] = dict()
    for word in doc.split():
        if word in bow_contagem[i]:
            bow_contagem[i][word] += 1
        else:
            bow_contagem[i][word] = 1
```

```python
bow_contagem
```

Bag of words ponderado 

```python

bow_ponderado = {}
N = {} # quantidade de termos em cada documento

for i,doc in enumerate(df['text']):
    bow_ponderado[i] = dict()
    N[i] = len(doc.split())
    for word in doc.split():
        if word in bow_ponderado[i]:
            bow_ponderado[i][word] += 1
        else:
            bow_ponderado[i][word] = 1

for doc in bow_ponderado:
    for word in bow_ponderado[doc]:
        bow_ponderado[doc][word] /= N[doc]
```

```python
bow_ponderado
```

Term Frequency

```python
import math
idf = {}
for word in vocab:
    idf[word] = 0
    #contabilizar em quantos documentos a palavra word aparece
    for doc in bow_contagem:
        if word in bow_contagem[doc]:
            idf[word] += 1

for word in idf:
    idf[word] = math.log(len(df['text'])/idf[word])
```

```python
idf
```

```python
bow_tfidf = {}

for doc in bow_ponderado:
    bow_tfidf[doc] = dict()
    for word in bow_ponderado[doc]:
        bow_tfidf[doc][word] = bow_ponderado[doc][word]*idf[word]
```

```python
bow_tfidf
```

```python
df_2 = pd.DataFrame().from_records(bow_tfidf).fillna(0)
df_2
```

```python
import math

def cosine_similarity(vetor1,vetor2):

    prod_interno = 0

    for i,valor in enumerate(vetor1):
        prod_interno += valor*vetor2[i]
    
    norma_vetor1 = 0
    for x in vetor1:
        norma_vetor1 += x*x
    norma_vetor1 = norma_vetor1**0.5

    #norma_vetor1 = math.sqrt(sum([x**2 for x in vetor1])) 

    norma_vetor2 = 0
    for x in vetor2:
        norma_vetor2 += x*x
    norma_vetor2 = norma_vetor2**0.5

    return prod_interno/(norma_vetor1*norma_vetor2)
```

```python
cosine_similarity(df_2['sentenca 1'], df_2['sentenca 1'])
```

```python
from operator import itemgetter
todas = []
for i, content_i in df_2.items():
  similarities=[]
  for j, content_j in df_2.items():
    if i != j:
      similarities.append({'index': i,'index_to_compare': j, 'cosine': cosine_similarity(content_i, content_j)})
      
  similarities = sorted(similarities, key=itemgetter('cosine'), reverse=True)
  todas.append(similarities[:10])

```

```python
all_top10 = pd.DataFrame()

for i, content in enumerate(todas):
  new_top10 = pd.DataFrame().from_dict(todas[i]).fillna(0)
  all_top10 = pd.concat([all_top10, new_top10])
all_top10
```
